```python
# Feature Engineering 
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]

print(df['Date'])
print(df['month'])
print(df['year'])
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey' ,axis = 1 ,inplace=True )
```

```python
df.info()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.info()
```

```python
df.head(2)
```

```python
df['arrival_hour'] = df['arrival_hour'].astype(int)
df['arrival_min'] = df['arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time' , axis= 1 , inplace= True)
```

```python
df.head(2)
```

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_hour'] = df['Departure_hour'].astype(int)
df['Departure_min'] = df['Departure_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time' , axis= 1 , inplace= True)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
}).fillna(1)  # Fill remaining NaN values with 1 as the mode
```

```python
df.isnull().sum()
```

```python
df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.drop('Route' , axis=1 , inplace= True)
```

```python
df.info()
```

```python
df['Duration_hour']=df['Duration'].str.split().str[0].str.split('h').str[0]
```

```python
df['Duration_minutes'] = df['Duration'].str.split(' ').str[1].str.replace('m', '').fillna('0')
```

```python
df['Duration'].isna().sum()
```

```python
df['Duration_minutes']
```

```python
df.head(2)
```

```python
# Separate hours and minutes from the 'Duration' column
df['Duration_hour'] = df['Duration'].str.extract('(\d+)h').fillna(0).astype(int)
df['Duration_minutes'] = df['Duration'].str.extract('(\d+)m').fillna(0).astype(int)

# Calculate total duration in minutes
df['Duration_total(mins)'] = df['Duration_hour'] * 60 + df['Duration_minutes']

# Check results
df.head(2)
```

```python
df['Duration'] = df['Duration_total(mins)']

df.head(2)
```

```python
df.rename(columns= {'Duration' : 'Duration(mins)'} , inplace=True)
df.drop('Duration_minutes' ,axis=1 , inplace=True)
df.drop('Duration_total(mins)' ,axis=1 , inplace=True)
df.drop('Duration_hour' ,axis=1 , inplace=True)
df.head(2)
```

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
df['Additional_Info'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline' , 'Source' , 'Destination']]).toarray()
```

```python
pd.DataFrame(encoder.fit_transform(df[['Airline' , 'Source' , 'Destination']]).toarray() , columns = encoder.get_feature_names_out())
```

```python
# Perform one-hot encoding
encoded_df = pd.DataFrame(encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray(), 
                          columns=encoder.get_feature_names_out())

# Concatenate the encoded DataFrame with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Optionally, you can drop the original columns if they're no longer needed
df.drop(columns=['Airline', 'Source', 'Destination'], inplace=True)

# Check the results
df.head()
```

```python
df.drop('Additional_Info' , axis = 1 , inplace= True)
```
