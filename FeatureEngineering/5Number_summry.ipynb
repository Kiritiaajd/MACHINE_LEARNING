## 5 NUMBER SUMMRY AND BOX PLOT 

```python
import numpy as np  
import pandas as pd   
```

```python
import random

lst_marks = [random.randint(0, 100) for _ in range(20)]
print(lst_marks)
```

```python
minimum , Q1 , median , Q3 , maximum = np.quantile(lst_marks ,[0 , 0.25 , 0.50 , 0.75 , 1.0])
```

```python
minimum , Q1 , median , Q3 , maximum
```

```python
IQR  = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5 * IQR
higher_fence = Q3 + 1.5 * IQR

print(lower_fence)
print(higher_fence)
```

```python
import seaborn as sns 
sns.boxplot(lst_marks)
  
```
