## Seaborn
Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating more attractive and informative statistical graphics. Seaborn simplifies the creation of complex visualizations by offering built-in themes, color palettes, and functions to handle data frames directly, especially when working with datasets from libraries like pandas.

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
## creare scatter plot 

import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill' , y = 'tip' , data = tips)
```

```python
sns.lineplot(x = 'size' , y = 'total_bill' , data = tips)
plt.title("Line Plotof Total Bill By size")
```

## Bar Plot
Categorical plots

```python

sns.barplot(x='day' , y = 'total_bill' , data = tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

## BOX PLOT 

```python
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips)
```

## VIOLION PLOT

```python
sns.violinplot(x = 'day' , y = 'total_bill', data = tips)
```

```python
## Histogram 

sns.histplot(tips['total_bill'] ,bins = 10, kde=True,)
```

```python
sns.kdeplot(tips['total_bill'] , fill= True)
```

```python
sns.pairplot(tips)
```

## Heat map

```python
corr = tips[['total_bill' , 'tip' , 'size']].corr()
corr
```

```python
sns.heatmap(corr , annot=True ,cmap = 'coolwarm' )
```
