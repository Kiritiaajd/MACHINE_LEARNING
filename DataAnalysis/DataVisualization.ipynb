## MATPLOTLIB

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is particularly popular for producing 2D plots such as line charts, bar charts, histograms, scatter plots, and more.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x= [1,2,3,4,5,6,7,8,9]
y = [9,8,7,6,5,4,3,2,1]

plt.plot(x ,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 100)  # Generates 100 points between -10 and 10

# Lambda function to calculate y = x^3
y = lambda x: x**3

# Apply lambda function to x values
y_values = y(x)

# Plot x and y
plt.plot(x, y_values)
plt.xlabel('x')
plt.ylabel('y = x^3')
plt.title('Plot of y = x^3')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 100)  # Generates 100 points between -10 and 10

# Define y = x^3 using a lambda function
y = x**3

# Create a customized line plot with red color and solid linestyle
plt.plot(x, y, color='red', linestyle='--')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y = x^3')
plt.title('Customized Plot of y = x^3')

# Show the plot
plt.show()
```

## Histogram


```python
import matplotlib.pyplot as plt

# Modified data for a better-looking histogram
data = [1, 2, 2, 3, 3, 4, 4, 5, 6, 7, 7, 8, 8, 9, 10, 10, 11, 12, 13, 14, 15, 15, 16, 17]

# Create the histogram with customizations
plt.hist(data, bins=5, color='red', edgecolor='black')

# Add title and labels
plt.title('Customized Histogram')
plt.xlabel('Data Values')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Scatter plot 

x = [1,2,3,4,5,6]
y = [7,6,8,4,5,7]
plt.scatter(x,y)
```
